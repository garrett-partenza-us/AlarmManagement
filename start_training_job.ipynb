```python
import sagemaker

sess = sagemaker.Session()
role = sagemaker.get_execution_role()
```

```python
from sagemaker.tensorflow import TensorFlow

tf_estimator = TensorFlow(entry_point='run_training_job.py', 
                          role=role,
                          instance_count=1, 
                          instance_type='ml.g4dn.2xlarge',
                          framework_version="2.2", 
                          py_version='py37',
                          script_mode=True,
                          hyperparameters={'epochs': 3, 'lookback':180, 'testcases': 100000, 'batch-size' : 32}
                         )
```

```python
tf_estimator.fit() 
```
